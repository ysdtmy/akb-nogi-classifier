```python
from bs4 import BeautifulSoup
import requests
import os
```

```python
imgs = []
PATH = './img/nogi'
URL = "https://48pedia.org/%E4%B9%83%E6%9C%A8%E5%9D%8246%E3%83%97%E3%83%AD%E3%83%95%E3%82%A3%E3%83%BC%E3%83%AB%E7%94%BB%E5%83%8F"
soup = BeautifulSoup(requests.get(URL).content,'lxml') # bsでURL内を解析

for link in soup.find_all("img"): # imgタグを取得しlinkに格納
    if link.get("src").endswith(".jpg"): # imgタグ内の.jpgであるsrcタグを取得
        imgs.append(link.get("src")) # imagesリストに格納

for i,img in enumerate(imgs):
    ext = img.split('/')[-1].split('.')[-1]
    re = requests.get('https://48pedia.org' + img)
    filecnt = len(os.listdir(PATH))
    filename = os.path.join(PATH,'nogi' + str(filecnt) + '.' + ext)
    print(filename)
    with open(filename, 'wb') as f:
        f.write(re.content)
```
